```python
tweets_df = pd.read_csv('/opt/anaconda3/lib/python3.9/site-packages/Giveme5W1H/data/source/bbc_breaking.csv')
```

```python
tweets_df['text'] = tweets_df['text'].str.replace("%", ' percent ', regex=False)
tweets_df['text'] = tweets_df['text'].str.strip()
tweets_df['text'] = tweets_df['text'].str.strip(":")
tweets_df['text'] = tweets_df['text'].str.replace('‚Äú', ' ', regex=False)
tweets_df['text'] = tweets_df['text'].str.replace('‚Äù', ' ', regex=False)
tweets_df['text'] = tweets_df['text'].str.replace('\"', ' ', regex=False)
tweets_df['text'] = tweets_df['text'].str.replace('‚Äò', '', regex=False)
tweets_df['text'] = tweets_df['text'].str.replace('\'', '', regex=False)
tweets_df['text'] = tweets_df['text'].str.replace('"', ' ', regex=False)
tweets_df = tweets_df.drop(['id', 'username'], axis=1)
```
